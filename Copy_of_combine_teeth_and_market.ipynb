```python
import tensorflow as tf
import numpy as np
import tensorflow.compat.v1 as tf
# import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

ONE_HOT_DIM1 = len(words1)

# function to convert numbers to one hot vectors
def to_one_hot_encoding(data_point_index,ONE_HOT_DIM):
    one_hot_encoding = np.zeros(ONE_HOT_DIM)
    one_hot_encoding[data_point_index] = 1
    return one_hot_encoding

X1 = [] # input word
Y1 = [] # target word

for x, y in zip(df1['input'], df1['label']):
    X1.append(to_one_hot_encoding(word2int1[ x ],ONE_HOT_DIM1))
    Y1.append(to_one_hot_encoding(word2int1[ y ],ONE_HOT_DIM1))

# convert them to numpy arrays
X_train1 = np.asarray(X1)
Y_train1 = np.asarray(Y1)

tf.compat.v1.disable_eager_execution()

# making placeholders for X_train and Y_train
x1 = tf.placeholder(tf.float32, shape=(None, ONE_HOT_DIM1))
y_label1 = tf.placeholder(tf.float32, shape=(None, ONE_HOT_DIM1))

# word embedding will be 2 dimension for 2d visualization
EMBEDDING_DIM = 2 

# hidden layer: which represents word vector eventually
W11 = tf.Variable(tf.random_normal([ONE_HOT_DIM1, EMBEDDING_DIM]))
b11 = tf.Variable(tf.random_normal([1])) #bias
hidden_layer = tf.add(tf.matmul(x1,W11), b11)

# output layer
W21 = tf.Variable(tf.random_normal([EMBEDDING_DIM, ONE_HOT_DIM1]))
b21 = tf.Variable(tf.random_normal([1]))
prediction1 = tf.nn.softmax(tf.add( tf.matmul(hidden_layer, W21), b21))

# loss function: cross entropy
loss1 = tf.reduce_mean(-tf.reduce_sum(y_label1 * tf.log(prediction1), axis=[1]))

# training operation
train_op1 = tf.train.GradientDescentOptimizer(0.05).minimize(loss1)
```

```python
#import tensorflow as tf
#import numpy as np
#import tensorflow.compat.v1 as tf
#tf.disable_v2_behavior()

ONE_HOT_DIM2 = len(words2)

# function to convert numbers to one hot vectors
def to_one_hot_encoding(data_point_index,ONE_HOT_DIM):
    one_hot_encoding = np.zeros(ONE_HOT_DIM)
    one_hot_encoding[data_point_index] = 1
    return one_hot_encoding

X2 = [] # input word
Y2 = [] # target word

for x, y in zip(df2['input'], df2['label']):
    X2.append(to_one_hot_encoding(word2int2[ x ],ONE_HOT_DIM2))
    Y2.append(to_one_hot_encoding(word2int2[ y ],ONE_HOT_DIM2))

# convert them to numpy arrays
X_train2 = np.asarray(X2)
Y_train2 = np.asarray(Y2)

# tf.compat.v2.disable_eager_execution()

# making placeholders for X_train and Y_train
x2 = tf.placeholder(tf.float32, shape=(None, ONE_HOT_DIM2))
y_label2 = tf.placeholder(tf.float32, shape=(None, ONE_HOT_DIM2))

# word embedding will be 2 dimension for 2d visualization
EMBEDDING_DIM = 2 

# hidden layer: which represents word vector eventually
W12 = tf.Variable(tf.random_normal([ONE_HOT_DIM2, EMBEDDING_DIM]))
b12 = tf.Variable(tf.random_normal([1])) #bias
hidden_layer = tf.add(tf.matmul(x2,W12), b12)

# output layer
W22 = tf.Variable(tf.random_normal([EMBEDDING_DIM, ONE_HOT_DIM2]))
b22 = tf.Variable(tf.random_normal([1]))
prediction = tf.nn.softmax(tf.add( tf.matmul(hidden_layer, W22), b22))

# loss function: cross entropy
loss2 = tf.reduce_mean(-tf.reduce_sum(y_label2 * tf.log(prediction), axis=[1]))

# training operation
train_op2 = tf.train.GradientDescentOptimizer(0.05).minimize(loss2)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
iteration = 20000
for i in range(iteration):
    # input is X_train which is one hot encoded word
    # label is Y_train which is one hot encoded neighbor word
    sess.run(train_op1, feed_dict={x1: X_train1, y_label1: Y_train1})
    if i % 3000 == 0:
        print('iteration '+str(i)+' loss is : ', sess.run(loss1, feed_dict={x1: X_train1, y_label1: Y_train1}))
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

iteration = 20000
for i in range(iteration):
    # input is X_train which is one hot encoded word
    # label is Y_train which is one hot encoded neighbor word
    sess.run(train_op2, feed_dict={x2: X_train2, y_label2: Y_train2})
    if i % 3000 == 0:
        print('iteration '+str(i)+' loss is : ', sess.run(loss2, feed_dict={x2: X_train2, y_label2: Y_train2}))
```
